```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Thu Jul  4 11:54:53 2019

@author: koushik
"""
#from config import fetch_file_configuration
from datetime import datetime 
import os
import time
import threading
   
def prRed(skk): 
    print("\033[91m {}\033[00m" .format(skk)) 

def prGreen(skk):
    print("\033[92m {}\033[00m" .format(skk)) 
    
def fetch_ftp_status(a_ftp_area,stn_code):
    print('In =',a_ftp_area,stn_code)
    before = dict ([(f, os.path.getsize(a_ftp_area+f)) for f in os.listdir (a_ftp_area) if stn_code in f and '.dat' in f]) # initial  list of files in the ftp area
    while True:
        time.sleep (10)
        after = dict ([(f, os.path.getsize(a_ftp_area+f)) for f in os.listdir (a_ftp_area) if stn_code in f and '.dat' in f]) #recent list of files in ftp area after 15 min
        added = [(f,os.path.getsize(a_ftp_area+f)>>10) for f in after if not f in before]
        if added: 
            message_summary = ' {} {} {} {} FTP File added  to :: {}                   Last updated---> {} {} {} {}  '
            prGreen(message_summary.format(u'\u2714',u'\u2714',u'\u2714',u'\u2714',stn_code,str(datetime.now())[:19],u'\u2714',u'\u2714',u'\u2714',u'\u2714'))
            for a_tuple in added :
                prGreen('FTP files Added to {}'.format(stn_code))
                message = '{} :: {} {}'
                prGreen(message.format(a_tuple[0],a_tuple[1],'Bytes'))
            
        if before == after:
            warning_message = '{} {} {} {} No FTP  Files added in the ::  {}                   Last updated---> {} {} {} {} ' 
            prRed(warning_message.format(u'\u2718',u'\u2718',u'\u2718',u'\u2718',stn_code,str(datetime.now())[:19],u'\u2718',u'\u2718',u'\u2718',u'\u2718'))
        before = after
        
if __name__ == "__main__": 
    
    directory_mounted_area= '/ftp2/'
    path_ftp_area_HSN = directory_mounted_area 
    path_ftp_area_SHG = directory_mounted_area 
    path_ftp_area_JPR = directory_mounted_area 
    path_ftp_area_BPL = directory_mounted_area 
    t1 = threading.Thread(target=fetch_ftp_status, args=(path_ftp_area_HSN,'HAS1',)) 
    t2 = threading.Thread(target=fetch_ftp_status, args=(path_ftp_area_SHG,'SHG',))
    t3 = threading.Thread(target=fetch_ftp_status, args=(path_ftp_area_JPR,'JPR',))
    t4 = threading.Thread(target=fetch_ftp_status, args=(path_ftp_area_BPL,'BPL',))
    t1.start()
    t2.start()
    t3.start()
    t4.start()
    t1.join()
    t2.join()
    t3.join()
    t4.join()
    print("Done!") 
  
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Thu Jul  4 11:54:53 2019

@author: koushik
"""
def prRed(skk): 
    print("\033[91m {}\033[00m" .format(skk)) 
def prGreen(skk):
    print("\033[92m {}\033[00m" .format(skk)) 


def fetch_file_configuration():
    configParser = configparser.RawConfigParser() 
    ensconfig_path = '/home/shift/Desktop/ens_status/'
    name_ensconfig = 'monitor_configfile.dat'
    ensconfig_path = ensconfig_path + name_ensconfig
    configParser.read(ensconfig_path)
    ens_directory = configParser.get('Configuration Settings','log_directory')
    ens_logfile = configParser.get('Configuration Settings','ens_logfile')
    return(ens_directory,ens_logfile)

import configparser
from time import sleep
import subprocess
from datetime import datetime
if __name__ == '__main__':
    OUTPUT_TEXT ='ENS_STATUS::{}{}{} TIME::{}'
    while True:
        ens_directory,ens_logfile = fetch_file_configuration()
        enslogfile = ens_directory + ens_logfile
        try:
            last_ens_status = subprocess.check_output(['tail', '-1', enslogfile])
            last_ens_status = last_ens_status.decode('utf-8')
            last_ens_status = last_ens_status.split(',')
            last_ens=(last_ens_status[1])
            if ' ENS : RUNNING' in last_ens:
                prGreen(OUTPUT_TEXT.format('UP',u'\u2714',u'\u2714',last_ens_status[0].split(':')[1]))
            elif 'ENS : NOT RUNNING\n' in last_ens:
                prRed(OUTPUT_TEXT.format('DOWN',u'\u2718',u'\u2718',last_ens_status[0].split(':')[1]))
            sleep(60)
        except FileNotFoundError :
            print('File not found...',datetime.now())
```
